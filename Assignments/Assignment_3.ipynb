# Check for Missing values and deal with them

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("abalone.csv")
print(df.shape)
print(df.info())
```

```python
df.describe()
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))
```

## Identifying Outliers with Visualization

```python
plt.boxplot(df["Rings"])
plt.show()
```

```python
df.boxplot(column='Rings', by='Length')
```

```python
df.Rings.hist()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.scatter(df['Rings'], df['Length'])
ax.set_xlabel('Rings')
ax.set_ylabel('Length')
plt.show()
```

```python
print(df['Rings'].quantile(0.10))
print(df['Rings'].quantile(0.90))
```

```python
df["Rings"] = np.where(df["Rings"] <2960.0, 2960.0,df['Rings'])
df["Rings"] = np.where(df["Rings"] >12681.0, 12681.0,df['Rings'])
print(df['Rings'].skew())
```

```python
index = df[(df['Length'] >= 100)|(df['Length'] <= 18)].index
df.drop(index, inplace=True)
df['Length'].describe()
```

```python
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```

```python
print(df['Rings'].quantile(0.50)) 
print(df['Rings'].quantile(0.95)) 
df['Rings'] = np.where(df['Rings'] > 325, 140, df['Rings'])
df.describe()
```

```python
df["Length"] = df["Rings"].map(lambda i: np.log(i) if i > 0 else 0) 
print(df['Length'].skew())
print(df['Rings'].skew())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("abalone.csv")
```

```python
df.head(10)
```

```python
columns = ['Length','Diameter','Height','Whole weight','Shucked weight','Rings']
```

```python
df = df.loc[:, columns]
df.head(10)
```

```python
features = ['Length','Diameter','Height','Whole weight','Shucked weight']
X = df.loc[:, features]
y = df.loc[:, ['Rings']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, train_size = .75)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor(max_depth = 2, random_state = 0)
```

```python
reg.fit(X_train, y_train)
```

```python
reg.predict(X_test[0:10])
```

```python
X_test.head(1)
```

```python
reg.predict(X_test.iloc[0].values.reshape(1,-1))
```

```python
score = reg.score(X_test, y_test)
print(score)
```
