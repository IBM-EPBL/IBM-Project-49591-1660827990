```python
#pdf and #cdf

counts,bin_edges=np.histogram(churn['Age'],bins=10,density=True)

plt.xlabel('Age')
pdf=counts/(sum(counts))
print("pdf=",pdf);
print("bin_edges=",bin_edges);
cdf=np.cumsum(pdf)
print("cdf=",cdf);
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf);
```
